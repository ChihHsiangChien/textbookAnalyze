```python
import csv
import pandas as pd
```

```python
conceptList = ['概論',
               '細胞',
               '養分',
               '運輸',
               '協調',
               '恆定',
               '生殖',               
               '遺傳',
               '演化',
               '分類',
               '生態',
               '環境', 
               '其他']
yearList = [57,61,72,83]
```

```python
df = pd.read_csv('課文節.csv', encoding='utf-8')
df_lab = pd.read_csv('實驗.csv', encoding='utf-8')
```

```python
with open('analyze.html', 'w', encoding='UTF-8') as myFile:
    myFile.write('<html>')
    myFile.write('<body>')
    myFile.write('<h1>')
    myFile.write('國民中學生物課程比較')    
    myFile.write('</h1>')
    myFile.write('<table border="1" style="width:80%; border-collapse: collapse; vertical-align:top" >')
    myFile.write('<tr style = "vertical-align:top">')
    myFile.write('<td>概念</td>')
    myFile.write('<td><p>57年1月1日(1968年)</p><p>國民中學暫行課程標準</p></td>')
    myFile.write('<td><p>61年10月(1972年)</p><p>國民中學課程標準</p></td>')
    myFile.write('<td><p>72年(1983年)</p><p>國民中學課程標準</p></td>')
    myFile.write('<td><p>72年(1983年)</p><p>國民中學課程標準<p><p>83年10月(1994年)修正</p><p>民國86年(1997年)教科書</p></td>')
        
    for concept in conceptList:
        
        #=================課文開始=======================
        myFile.write('<tr><td>' + concept + '</td>')
                     
        for year in yearList:
            cond1 = df['概念'] == concept
            cond2 = df['課程標準'] == year
            print(concept)
            
            ##print(year)
            sectionList = df[cond1 & cond2][['冊','章號','節號','節名']].values.tolist() 
            myFile.write('<td style = "vertical-align:top" >')
            for section in sectionList:
                print(section[0])
                if section[0] == 1 :
                    book = '上' 
                else:
                    book = '下'
                item = book + ' ' + str(section[1])+'-'+ str(section[2]) + ' ' + section[3]
                print(item)                     
                
                myFile.write('<p>')
                myFile.write( item )                     
                myFile.write('</p>')                     

            myFile.write('</td>')
        myFile.write('</tr>')    
        
        #=================課文結束=======================
        
        #=================實驗或活動開始=======================

        myFile.write('<tr style = "background-color: #F0F0F0;" ><td>' + concept +'<p><b>實驗活動</b></p>' '</td>') 
        for year in yearList:
            cond1 = df_lab['概念'] == concept
            cond2 = df_lab['課程標準'] == year
            print(concept)
            
            ##print(year)
            labList = df_lab[cond1 & cond2][['冊','編號','名稱']].values.tolist() 
            myFile.write('<td style = "vertical-align:top" >')
            for lab in labList:
                if labList[0] == 1 :
                    book = '上' 
                else:
                    book = '下'
                item = book + ' ' + str(lab[1])+' '+ lab[2]
                print(item)                     
                
                myFile.write('<p>')
                myFile.write( item )                     
                myFile.write('</p>')                     

            myFile.write('</td>')
        myFile.write('</tr>')    
        #=================實驗或活動結束=======================        
        

        
        
    myFile.write('</tr>')
    myFile.write('</table>')
    myFile.write('</body>')
    myFile.write('</html>')


```
