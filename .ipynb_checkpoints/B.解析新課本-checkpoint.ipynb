```python
import csv
```

```python
with open('108.txt',encoding ='utf-8') as f:
    lines = f.readlines()
```

```python
lines
```

```python
text_list = []
chapter_num = 1
section_num = 1
for line in lines:
    if '章' in line:
        new_line = line.split('\t')
        chapter = new_line[1].replace('\n', '').replace('\r', '')
        chapter_num += 1
        section_num = 1
        #print(chapter)
        if '第一章' in line:
            chapter_num = 1
    if '第' in line and '節' in line:
        new_line = line.split('\t')
        section = new_line[1].replace('\n', '').replace('\r', '')
        text = [chapter_num,chapter,section_num,section]
        print(text)
        section_num += 1
        text_list.append(text)
    
```

```python
with open('108clean.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(text_list)
```

```python
text_list = []
chapter_num = 1
section_num = 0
for line in lines:
    if '章' in line:
        new_line = line.split('\t')
        chapter = new_line[1].replace('\n', '').replace('\r', '')
        chapter_num += 1
        section_num = 1
        #print(chapter)
        if '第一章' in line:
            chapter_num = 1
    if '第' in line and '節' in line:
        new_line = line.split('\t')
        section = new_line[1].replace('\n', '').replace('\r', '')
        section_num += 1
    if '活動' in line:
        lab = line.split('\t')
        lab[1] = lab[1].replace('活動', '')
        lab[2] = lab[2].replace('\n', '')
        text = [chapter_num,section_num-1,lab[1],lab[2]]
        print(text)
        text_list.append(text)
print(text_list)
```

```python
with open('108clean_lab.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(text_list)
```
